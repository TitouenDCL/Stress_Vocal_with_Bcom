```python
import numpy as np
import matplotlib
import scipy
import tqdm
import sklearn
import plotly
import pyAudioAnalysis
from pyAudioAnalysis.audioTrainTest import random_split_features
from sklearn.model_selection import train_test_split
from pyAudioAnalysis import audioBasicIO
from pyAudioAnalysis import MidTermFeatures
import matplotlib.pyplot as plt
import train_test
import train_test_actor
from pyAudioAnalysis import audioTrainTest as aT

def accuracy(cm):
    return (cm[0,0]+cm[1,1])/(cm[0,0]+cm[1,1]+cm[0,1]+cm[1,0])
```

Creation de tous les modèle de gradientboosting

# On va maintenant utiliser une autre méthode de séparation des bases de données 
# On isole un acteur ( qui a fait des audios stress et other ) sur lequel on n'entraine pas le modèle, puis on utilise cet acteur pour faire des tests

# On a commencé avec la méthode de calcul "gradientboosting" qui donne les meilleures au premiers test
# On répète le procédé pour 9 acteurs differents

# Création des modèles pour les acteurs de 11 à 19 (il y a 24 acteurs au total)

```python
for i in range(11, 20):
    print(str(f"{i}"))
    train_test_actor.data_split_actor(f"{i}", 'DataBase/Stress/RAVDES', 'DataBase/other/RAVDES', f"DataBase/Random_Train_Test_Actors_{i}")
    print(f"split")
    aT.extract_features_and_train([f"DataBase/Random_Train_Test_Actors_{i}/train\\fear",
                               f"DataBase/Random_Train_Test_Actors_{i}/train\\other"],
                              1.0, 1.0, aT.shortTermWindow, aT.shortTermStep,
                              "gradientboosting",
                              f"Model_Fear_gradientboosting_Actors_{i}")
```

# puis on affiches les résultat d'accuracy

```python
mean_accuracy=0
for i in range(11, 20):
    cm, thr_prre, pre, rec, thr_roc, fpr, tpr = aT.evaluate_model_for_folders(
        [f"DataBase/Random_Train_Test_Actors_{i}/test\\fear",
         f"DataBase/Random_Train_Test_Actors_{i}/test\\other"],
        f"Model_Fear_gradientboosting_Actors_{i}",
        "gradientboosting",
        "fear")
    mean_accuracy += accuracy(cm)
    print(accuracy(cm))
mean_accuracy = mean_accuracy/9
print(f"{mean_accuracy=}")
```
